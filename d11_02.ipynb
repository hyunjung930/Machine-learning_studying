```python
import matplotlib.pyplot as plt
```

```python
from numpy import square
def draw_fruits(arr_3d): # 3d: 3차원 구조
  ''' 
  arr_3d.shape: (sample #, image's width, image's height)
  '''
  n = len(arr_3d) #샘플의 수
  rows = int(np.ceil(n/10)) #np.ceil은 float 형식으로 반환으로 int로 바꿔줘야한다.
  cols = n if rows < 2 else 10
  fig, axes = plt.subplots(rows,cols, squeeze = False)
  for i in range(rows):
    for j in range(cols):
      if i*10 + j < n:
        axes[i,j].imshow(arr_3d[i *10 + j], cmap ='gray_r')
      axes[i, j].axis('off')
  plt.show()
```

```python
#주성분 분석
from sklearn.decomposition import PCA

pca = PCA(n_components = 50)
pca.fit(fruits_2d)
```

```python
pca.components_.shape #10000개 특성, 50개의 이미지
```

```python
pca.components_.reshape(-1,100,100).shape #3차원 배열 변경
```

```python
draw_fruits(fruits)
```

```python
draw_fruits(pca.components_.reshape(-1,100,100))
```

```python
fruits_pca = pca.transform(fruits_2d)
fruits_pca.shape #50개로 변경 
```

```python
fruits_inverse = pca.inverse_transform(fruits_pca)
fruits_inverse.shape
```

```python
draw_fruits(fruits_inverse.reshape(-1,100,100))
```

```python
pca.explained_variance_# variance는 분산값
```

```python
pca.explained_variance_ratio_ #ratio
```

```python
plt.plot(pca.explained_variance_ratio_)
plt.xlabel('component #')
plt.ylabel('ratio')
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
target = np.array([0]*100 +[1]*100 +[2]*100)
np.unique(target, return_counts = True)
```

```python
from sklearn.model_selection import cross_validate

scores = cross_validate(lr, fruits_2d, target)
print(f"calidation score: {np.mean(scores['test_score'])}")
print(f"fit time: {np.mean(scores['fit_time'])}")
```

```python
fruits_pca.shape
```

```python
scores = cross_validate(lr, fruits_pca, target)
print(f"calidation score: {np.mean(scores['test_score'])}")
print(f"fit time: {np.mean(scores['fit_time'])}")
```

```python
pca = PCA(n_components = 0.55)
pca.fit(fruits_2d)
pca.components_.shape
```

```python
draw_fruits(pca.components_.reshape(-1,100,100))
```

```python
fruits_pca = pca.transform(fruits_2d)
fruits_pca.shape
```

```python
scores = cross_validate(lr, fruits_pca, target)
print(f"calidation score: {np.mean(scores['test_score'])}")
print(f"fitted time: {np.mean(scores['fit_time'])}")
```

```python
from sklearn.cluster import KMeans

km = KMeans(n_clusters=3, random_state=42)
km.fit(fruits_pca)
np.unique(km.labels_, return_counts =  True)
```

```python
for label in range(3):
  data = fruits_pca[km.labels_ == label]
  plt.scatter(data[:,0],data[:,1])
  plt.plot(km.cluster_centers_[label,0],km.cluster_centers_[label,1], marker ='+')
  plt.legend(['pinapple','banana','apple'])
plt.show()
```
