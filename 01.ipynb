```python
# *,** 예제
a = [1,2,3]
b = {"x":1,"y":2}
```

```python
# positional, keywords
def fn (*args, **kwargs):
  for i, arg in enumerate(args): #enumerate: return을 튜플형태로 만들어줌(index형태로 만들어 주는 것)
    print(f"{i} : {arg}")
  for k, v in kwargs.items():
    print(f"{k} : {v}") # k: key / v: value
    #positional args 
```

```python
fn(*a,**b) #unpack
```

```python
fn(1,2,3,x=1,y=2)
```
